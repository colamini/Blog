2、二次函数

一般式： $y=ax^2+bx+c (a!=0)$ 

顶点式： $y=a(x-m)^2+n (a!=0)$

定义域：R

对称轴：$-\frac{b}{2a}$

最大/小值： $\frac{4ac-b^2}{4a}$

应用场景：

1、经济学中的投资分析、销售定价
2、竞技体育中的篮球落点预测
3、复杂函数的二阶近似
4、很多优化算法中


```python
import matplotlib.pyplot as plt
import numpy as np

# make data
x1 = np.linspace(0, 100, 100) # 1-10之间，生成100个数
y1 = np.sqrt(2 * x1) 
y2 = -np.sqrt(2 * x1) 


# plot
plt.figure()

plt.plot(x1, y1, linewidth=2.0)
plt.plot(x1, y2, linewidth=2.0)

plt.show()
```

3、椭圆方程
$\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$

一般式为：
$AX^2+By^2+Cxy+Dx+Ey+F=0$


```python
import matplotlib.pyplot as plt
import numpy as np

# make data
a = 2
b = 5
x1 = np.linspace(-2,2,100) # 1-10之间，生成100个数

y1 = np.sqrt(abs((pow(a,2) * pow(b,2) - pow(b,2) * pow(x1,2)) / pow(a, 2))) 
y2 = -np.sqrt(abs((pow(a,2) * pow(b,2) - pow(b,2) * pow(x1,2)) / pow(a, 2))) 


# plot
plt.figure()
plt.plot(x1, y1, linewidth=2.0)
plt.plot(x1, y2, linewidth=2.0)

plt.show()
```

4、心形线

```python
x_data1 = np.linspace(-1,1,1000) 
# 根据心形公式，得到y的表达式，由于有正负之分，分开表示
def f(x):
    y_data1 = np.sqrt(1 - x**2) + pow(np.abs(x),float(2)/float(3))
    y_data2 = -np.sqrt(1 - x**2)+ pow(np.abs(x),float(2)/float(3))
    return y_data1, y_data2

# 开始画图
plt.figure()
plt.plot(x_data1, f(x_data1)[0])
plt.plot(x_data1, f(x_data1)[1])

plt.show()
```

 5、Sigmoid函数

 $ y=\frac{1}{1+ e^{-x}} $


多元Sigmoid

 $ h(x_1, x_2) =\theta_0 + \theta_1 X_1 + \theta_2 X_2+ \theta_3 X_1^2 + \theta_4 X_2^2 + \theta_5 X_1 X_2 = 0 $
 
 $ y=\frac{1}{1+ e^{-h(x_1, x_2)}} $

```python
import matplotlib.pyplot as plt
import numpy as np

# make data
x = np.linspace(-10, 10, 1000) # 1-10之间，生成100个数
y = 1 / (1 + pow(np.e, -x))

# plot
plt.figure()
plt.plot(x, y, linewidth=2.0)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
import pandas as pd
import matplotlib.ticker as ticker

def g(x):
    return 1 / (1 + pow(np.e, -x))

X1 = np.linspace(-1, 1, 10)
X2 = np.linspace(-1, 1, 10)

A1 = g(-10 * X1 - 10 * X2 + 5)
A2 = g(10 * X1 + 10 * X2 - 15)

X1_new, X2_new = np.meshgrid(X1, X2)
A1_new, A2_new = np.meshgrid(A1, A2)

Y = g(10 * A1_new + 10* A2_new - 5)

print(Y)
# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.set_title('x1 XNOR x2')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
# ax.plot_trisurf(X1_new, X2_new, Y)
# plt.plot(X1_new, X2_new, 'b--')
ax.plot_wireframe(X1_new, X2_new, Y)


plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
import pandas as pd


# Make data
X1 = np.linspace(-1, 1, 100)
X2_1 = np.sqrt(abs(1 - pow(X1, 2)))
X2_2 = -np.sqrt(abs(1 - pow(X1, 2)))

X1_new = np.linspace(-3, 3, 100)
X2_new = np.linspace(-3, 3, 100)
X1_new, X2_new = np.meshgrid(X1_new, X2_new)

Y = 1 / (1 + pow(np.e, -(pow(X1_new, 2)  + pow(X2_new, 2) - 1)))

plt.style.use('_mpl-gallery')

print(Y)
# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
plt.plot(X1, X2_1)
plt.plot(X1, X2_2)

# ax.plot_trisurf(X1_new, X2_new, Y)
ax.plot_surface(X1_new, X2_new, Y)

plt.show()
```
