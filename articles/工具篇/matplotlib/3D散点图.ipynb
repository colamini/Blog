```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import torch

# tensor3D
tensor = torch.linspace(1, 30, steps=30).view(3,2,5)
# NumPy
array = tensor.numpy()

# 
depth, height, width = array.shape

# 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 
x, y, z = np.indices((width, height, depth))

# 
x = x.flatten()
y = y.flatten()
z = z.flatten()
values = array.flatten()

# 
ax.scatter(x, y, z, c=values, cmap='viridis')

# 
ax.set_xlim([0, 5])
ax.set_ylim([-2, 5])
ax.set_zlim([-2, 5])

# 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 
cbar = plt.colorbar(ax.collections[0])
cbar.set_label('Value')

# 
plt.show()
```
