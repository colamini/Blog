```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import torch

# 假设您有一个名为tensor的3D张量对象
tensor = torch.linspace(1, 30, steps=30).view(3,2,5)
# 将张量转换为NumPy数组
array = tensor.numpy()

# 获取数组的形状
depth, height, width = array.shape

# 创建3D图形对象
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 生成坐标点
x, y, z = np.indices((width, height, depth))

# 展平坐标和数值
x = x.flatten()
y = y.flatten()
z = z.flatten()
values = array.flatten()

# 绘制散点图
ax.scatter(x, y, z, c=values, cmap='viridis')

# 设置坐标轴范围
ax.set_xlim([0, 5])
ax.set_ylim([-2, 5])
ax.set_zlim([-2, 5])

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 添加颜色条
cbar = plt.colorbar(ax.collections[0])
cbar.set_label('Value')

# 显示图形
plt.show()
```
